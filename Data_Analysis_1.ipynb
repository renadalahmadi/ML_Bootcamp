```python
import pandas as rData
import matplotlib.pyplot as myPlot
```

```python
url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx"
myData = rData.read_excel(url,
                          sheet_name = "Canada by Citizenship",
                          skiprows = range(20),
                          skipfooter = 2)
myData
```

```python
# the first 5 rows 
myData.head(5)
# the last 5 rows 
myData.tail(7)
```

```python
# columns names 
myData.columns.values
# check size of Dataframe
myData.shape
```

```python

# checking Data types in the Dataset
print(myData.dtypes)
# checking Data statistics in the Dataset
print(myData.describe())
# checking Data general information about Dataset
print(myData.info())
# checking missing values 
print(myData.isnull().sum())
```

```python
myData.rename(columns= {'OdName':'Country', 'AreaName': 'Continent', 'RegName':'Region'}, inplace = True)
```

```python
myData.drop(["Coverage", "AREA", "REG", "DEV", "DevName"], axis = 1, inplace = True)
```

```python
print(myData.iloc[0:2])
print(myData["Region"].loc[0:20])
```

```python
myData["Total"] = myData.sum(axis = 1)
myData
```

```python
print(myData)
myData.set_index('Country', inplace = True)
myData.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
Data5 = myData.head(5)
Data5 = Data5[range(1980, 2014)].transpose()
Data5.plot(kind = "area")
myPlot.title('Immigration form Top 5 Countries')
myPlot.ylabel('Number of Immigrant')
myPlot.xlabel('Years')
myPlot.show()
```

```python
# Export dataset 
myData.to_csv("path.csv")
```
